### Task - Find the most popular Trios-I
- Your a producer looking to make a blockbuster movie. there will primarily be three lead roles in your movie and you wish to cast the most popular actors for it, now since you dont want to take a risk, you will cast a trio which has already acted in together in a movie before. the metric that you have chosen to check the popularity is the facebook likes of each of these actors.

##Task - find the most popular trios- II
- In the previous subtask find out the popular trio based on the total number of facebook likes, Lets add a small condition to it and make sure that all three actors are popular.
- The condition is none of the three actors, facebook likes should be less than half of the other two.

- e.g the following is a valid combo
- - act1_fb_likes = 70000
- - act_2_fb_likes = 40000
- -act_3_fb_likes = 50000

- but the below one is not.
- - act_1_fb_likes = 70000
- - act_2_fb_likes = 40000
- - act_3_fb_likes = 30000.


- Check how many of that trios are satisying the condition.

- Which is the most popular trio after applying the condition above??

**Even though you are finding this out by a natural inspection of the dataframe,can you also achieve this through some-if-else statements to incorporate this.**


```python
act1 = df['actor_1_facebook_likes']/2
act2 = df['actor_2_facebook_likes']/2
act3 = df['actor_3_facebook_likes']/2

a = ((df['actor_1_facebook_likes'] > act1) & (df['actor_1_facebook_likes'] > act3))
b = ((df['actor_2_facebook_likes'] > act2) & (df['actor_2_facebook_likes'] > act3))
c = ((df['actor_3_facebook_likes'] > act3) & (df['actor_3_facebook_likes'] > act2))
eligible = a & b & c

df['eligible'] = eligible
df.loc[eligible, ['eligible', 'actor_1_name', 'actor_2_name', 'actor_3_name']].head(5)
```

```python
df_by_genre = df_by_genre.apply(pd.to_numeric, errors='ignore')
```

```python
#Group the movies by individual genres
df_by_g1 = df_by_genre.groupby('genre_1').sum()
df_by_g2 = df_by_genre.groupby('genre_2').sum()
df_by_g3 = df_by_genre.groupby('genre_3').sum()
```

```python
# Convert all columns to numeric type, coercing errors to NaN
df_by_g1 = df_by_g1.apply(pd.to_numeric, errors='coerce')
df_by_g2 = df_by_g2.apply(pd.to_numeric, errors='coerce')
df_by_g3 = df_by_g3.apply(pd.to_numeric, errors='coerce')

# Perform the addition with fill_value=0
df_add = df_by_g1.add(df_by_g2, fill_value=0).add(df_by_g3, fill_value=0)
```
